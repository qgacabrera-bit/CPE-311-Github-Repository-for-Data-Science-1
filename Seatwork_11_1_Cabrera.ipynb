```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def solve_linear(m, x, b):
  y = b
  results = []
  for i in range(len(m)):
    y += m[i] * x[i]
    results.append(y)


  return y, results

print("Part 1")
weights = [3]
x_values = [0]
intercept = 1

result = solve_linear(weights, x_values, intercept)

print("Weights (m):", weights)
print("X values:", x_values)
print("Intercept (b):",intercept)
print("Calculated linear regression value (y):", result[0])

print("Graph")
plt.scatter(x_values, result[1])
plt.xlabel('Input Feature Value')
plt.ylabel('Results')
plt.title('Input Values vs. Results ')
plt.grid(True)
plt.show()
```

Insight: since there is only 1 x value and weight, we can only see one output value in the graph.

```python
print("Part 2: Additional Example 1")
weights = [3,-2,4]
x_values = [3,-1,2]
intercept = 1

result = solve_linear(weights, x_values, intercept)

print("Weights (m):", weights)
print("X values:", x_values)
print("Intercept (b):",intercept)
print("Calculated linear regression value (y):", result[0])

print("Graph")
plt.plot(x_values, result[1], '-o')
plt.xlabel('Input Feature Value')
plt.ylabel('Results')
plt.title('Input Values vs. Results ')
plt.grid(True)
plt.show()
```

Insight: Even though the input value is lower, the results value still goes up since the weights and value in the portion are both negative and gets added to the previous value.


```python
print("Part 2: Additional Example 2")
weights = [3,-2,4]
x_values = [10,0,-10]
intercept = 1

result = solve_linear(weights, x_values, intercept)

print("Weights (m):", weights)
print("X values:", x_values)
print("Intercept (b):",intercept)
print("Calculated linear regression value (y):", result[0])

print("Graph")
plt.plot(x_values, result[1], '-o')
plt.xlabel('Input Feature Value')
plt.ylabel('Results')
plt.title('Input Values vs. Results ')
plt.grid(True)
plt.show()
```

Insight: Similar to the previous value, we can see that the result is sensitive to the high input values, while the 2 input values are the same, the output values are different because of the weights.
