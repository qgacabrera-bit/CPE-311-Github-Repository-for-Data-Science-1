#### **Exercise 2**
- With faang, use type conversion to change the date column into a datetime and the volume column into integers. Then, sort by date and ticker. 
- Find the seven rows with the highest value for volume. 
- Right now, the data is somewhere between long and wide format. Use melt() to make it completely long format. Hint: date and ticker are our ID variables (they uniquely identify 
each row). We need to melt the rest so that we don't have separate columns for open, high, low, close, and volume. 


```python
faang['date'] = pd.to_datetime(faang['date'])
faang['volume'] = faang['volume'].astype(int)
faang = faang.sort_values(by=['date','ticker'])
faang
```

```python
top7 = faang.nlargest(7,'volume')
top7
```

```python
faang_melted = faang.melt(
    id_vars=['date','ticker'],
    value_vars=['open','high','low','close','volume'],
    var_name='metric',
    value_name= 'value'
)
faang_melted
```

#### **Exercise 3** 
- Using web scraping, search for the list of the hospitals, their address and contact information. Save the list in a new csv file, hospitals.csv. 
- Using the generated hospitals.csv, convert the csv file into pandas dataframe. Prepare the data using the necessary preprocessing techniques. 

```python
import requests
from bs4 import BeautifulSoup
```

```python
headers = {    
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0 win64; x64)"
}

url = 'https://quezoncity.gov.ph/covid-19-watch/hospitals-and-health-centers-directory/'

response = requests.get(url, headers=headers)

print(response.status_code)
print(response.text)
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
print(soup.find('title').text)
```

```python
data = soup.find(class_='advgb-tab-_cec1d9-8a advgb-tab-body')
```

```python
print(data.prettify())
```

```python
names = []
addresses = []
contacts = []

def get_text_or_na(element):
    return element.get_text(strip=True) if element else "Not available"

boxes = data.select('.healthcenter-box')

for box in boxes:
    names.append(get_text_or_na(box.select_one('.healthcenter-title')))
    addresses.append(get_text_or_na(box.select_one('i.dashicons-location + span')))
    contacts.append(get_text_or_na(box.select_one('i.dashicons-phone + span')))

df = pd.DataFrame({
    "Name": names,
    "Address": addresses,
    "Contact": contacts,
})

df.to_csv("health_centers.csv", index=False, encoding="utf-8")
df
```
