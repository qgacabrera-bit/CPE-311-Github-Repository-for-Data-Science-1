```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
customer = pd.read_json('/content/drive/MyDrive/3rd Year Subs/Dat Sci 1/Midterm Skills Exam/customers_data.json')
customer = pd.DataFrame(customer)
customer
```

```python
sales = pd.read_csv('/content/drive/MyDrive/3rd Year Subs/Dat Sci 1/Midterm Skills Exam/sales_data_raw.csv')
sales
```

```python
sales['TransactionDate'] = pd.to_datetime(sales['TransactionDate'])
sales
```

```python
customer['JoinDate'] = pd.to_datetime(customer['JoinDate'])
customer
```

```python
sales['FinalTransactionAmount'] = sales['Quantity'] * sales['Price'] - sales['Discount']
sales
```

```python
customer_sales = pd.merge(customer, sales, left_on='CustomerID', right_on='CustomerID')
customer_sales.drop(columns=['JoinDate'], inplace = True)
customer_sales
```

```python
sales.duplicated()
sales.drop_duplicates()
```

```python
final_trans = customer_sales.groupby('CustomerID')['FinalTransactionAmount'].sum()
plt.bar(final_trans.index, final_trans.values)
plt.xlabel('Customer ID')
plt.ylabel('Total Amount (in USD)')
plt.title('Total Amount Spent per Customer ID')
plt.show()
```

```python
total_discount = customer_sales.groupby('CustomerID')['Discount'].sum()
plt.bar(total_discount.index, total_discount.values)
plt.xlabel('Customer ID')
plt.ylabel('Total Discount (in USD)')
plt.title('Total Savings per Customer ID')
plt.show()
```

```python
most_bought = customer_sales.groupby('Product')['Quantity'].sum()
plt.bar(most_bought.index, most_bought.values)
plt.xlabel('Product')
plt.ylabel('Total Quantity Sold')
plt.title('Total Quantity Sold per Product')
```

```python
most_sales = customer_sales.groupby('Product')['FinalTransactionAmount'].sum()
plt.bar(most_sales.index, most_sales.values)
plt.xlabel('Product')
plt.ylabel('Total Sales (in USD)')
plt.title('Total Sales per Product')
```

```python
most_spent = customer_sales.groupby('CustomerID')['FinalTransactionAmount'].max()
plt.bar(most_spent.index, most_spent.values)
plt.xlabel('Customer ID')
plt.ylabel('Amount Spent (in USD)')
plt.title('Most Amount Spent per Customer in a Single Transaction')
plt.show()
```
