1. Create a code that demonstrate how to use recursion method to solve problem

```python
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n-1)

x = int(input("Enter a number to get a factorial of: "))
print("Factorial of", x, "is",factorial(x))
```

2. Create a program codes that demonstrate how to use dynamic programming to solve the same problem 

```python
def factorial_dp(n):
  table =  [-1 for i in range(n+1)]
  table[0] = 1
  table[1] = 1
  i = 2
  while i <= n:
    table[i] = i*table[i-1]
    i+=1
  return table[n]
y = int(input("Enter a number to get a factorial of: "))
print("Factorial of", y, "is",factorial_dp(y))
```

##### Question: 
Explain the difference of using the recursion from dynamic programming using the given sample codes to solve the same problem

The recursive approach calculates the factorial by having the function call itself with smaller numbers until it reaches 0. For example, factorial(4) calls factorial(3), factorial(2), and so on, then multiplies the results on the way back. This method is simple and easy to understand but can be slow and use a lot of memory for large numbers. The dynamic programming approach, on the other hand, uses a table to store the factorials of all numbers from 0 to n and calculates them one by one. This way, each factorial is calculated only once, making it faster and safer for large numbers. In short, recursion is simple but less efficient, while dynamic programming is more efficient and avoids repeated work.

* Analyze three different techniques to solve knapsacks problem
1. Recursion
2. Dynamic Programming
3. Memoization

```python
#sample code for knapsack problem using recursion
def rec_knapSack(w, wt, val, n):

  #base case
  #defined as nth item is empty;
  #or the capacity w is 0
  if n == 0 or w == 0:
    return 0

  #if weight of the nth item is more than
  #the capacity W, then this item cannot be included
  #as part of the optimal solution
  if(wt[n-1] > w):
    return rec_knapSack(w, wt, val, n-1)

  #return the maximum of the two cases:
  # (1) include the nth item
  # (2) don't include the nth item
  else:
    return max(
        val[n-1] + rec_knapSack(
            w-wt[n-1], wt, val, n-1),
            rec_knapSack(w, wt, val, n-1)
    )
```

```python
#To test:
val = [60, 100, 120] #values for the items
wt = [10, 20, 30] #weight of the items
w = 50 #knapsack weight capacity
n = len(val) #number of items

rec_knapSack(w, wt, val, n)
```

```python
#Dynamic Programming for the Knapsack Problem
def DP_knapSack(w, wt, val, n):
  #create the table
  table = [[0 for x in range(w+1)] for x in range (n+1)]

  #populate the table in a bottom-up approach
  for i in range(n+1):
    for w in range(w+1):
      if i == 0 or w == 0:
        table[i][w] = 0
      elif wt[i-1] <= w:
        table[i][w] = max(val[i-1] + table[i-1][w-wt[i-1]],
                          table[i-1][w])
  return table[n][w]
```

```python
#To test:
val = [60, 100, 120]
wt = [10, 20, 30]
w = 50
n = len(val)

DP_knapSack(w, wt, val, n)
```

```python
#Sample for top-down DP approach (memoization)
#initialize the list of items
val = [60, 100, 120]
wt = [10, 20, 30]
w = 50
n = len(val)

#initialize the container for the values that have to be stored
#values are initialized to -1
calc =[[-1 for i in range(w+1)] for j in range(n+1)]


def mem_knapSack(wt, val, w, n):
  #base conditions
  if n == 0 or w == 0:
    return 0
  if calc[n][w] != -1:
    return calc[n][w]
  
  #compute for the other cases
  if wt[n-1] <= w:
    calc[n][w] = max(val[n-1] + mem_knapSack(wt, val, w-wt[n-1], n-1),
                     mem_knapSack(wt, val, w, n-1))
    return calc[n][w]
  elif wt[n-1] > w:
    calc[n][w] = mem_knapSack(wt, val, w, n-1)
    return calc[n][w]

mem_knapSack(wt, val, w, n)
```

**<h3>Code Analysis</h3>**
<hr>
<h4>Recursion Approach</h4>

-  Performance: This approach works well for small numbers of items. However, it becomes very slow and inefficient as the number of items grows, because the same subproblems are recalculated multiple times. For example, in this activity, even though there were only three items, recursion already had multiple repeated calculations. With larger inputs, execution time can increase dramatically.


-  Usability: Recursion is easy to write and understand for beginners. It is a good way to learn the problem-solving logic behind the knapsack problem. You can quickly implement it without worrying about tables or extra storage.


-  Readability: The code is short and straightforward, making it very easy to follow. It clearly shows the logic of checking whether to include an item or not. Beginners can easily trace the recursive calls to understand how the solution is formed.



<h4>Dynamic Programming</h4>

-  Performance: This is the fastest and most reliable method among the three. Because each subproblem is solved only once and stored in the table, the algorithm can handle larger numbers of items and capacities efficiently. It avoids the exponential time complexity of recursion.


-  Usability: Dynamic programming is very practical for real-life applications. It works well when you need reliable and predictable performance, such as inventory management, resource allocation, or finance-related problems.


-  Readability: The code is longer and uses nested loops, which may seem more complex to beginners. However, once you understand the structure of the table, it is logical and easy to follow. You can see how each cell of the table represents the solution to a subproblem.



<h4>Memoization</h4>


-  Performance: Memoization is much faster than pure recursion because each subproblem is calculated only once. It performs almost as well as bottom-up dynamic programming, especially for medium-sized datasets.


-  Usability: This method is helpful if you already have a recursive solution and want to optimize it without rewriting everything. It is particularly useful when the recursive logic is already easy to understand, but execution time needs improvement.


-  Readability: The code remains easy to read because it keeps the recursive structure. Adding a memoization table introduces some extra complexity, but it is not difficult to understand. You can still follow the logic of including or excluding items recursively.

Task 2: Create a sample program that find the nth number of Fibonacci Series using Dynamic Programming

```python
def fib(n, res):
    if n == 1:
        return 0
    if n == 2:
        return 1

    if n not in res:
        res[n] = fib(n-1,res) + fib(n-2, res)
    return res[n]

x = int(input("Enter a number in the fibonacci sequence: "))
print("First",x,"numbers in the fibonacci sequence:")
for i in range(1, x+1):
    print(fib (i, {}), end=' ')
```

A project manager was given a list of project proposals to be implemented in the upcoming months, he wants to find out which of the projects will fit the limited budget while still acquiring the maximum profit.

```python
costs = [20, 30, 40, 50]       # Project costs
profit = [60, 100, 120, 150]  # Profit/Kita
budget = 100
n = len(costs)

def budget_rec(costs, returns, budget, n):
    # Base case: no projects or no budget left
    if n == 0 or budget == 0:
        return 0
    
    # If project cost exceeds budget, skip it
    if costs[n-1] > budget:
        return budget_rec(costs, returns, budget, n-1)
    
    #Include the project
    include = returns[n-1] + budget_rec(costs, returns, budget - costs[n-1], n-1)
    #Exclude the project
    exclude = budget_rec(costs, returns, budget, n-1)
    
    return max(include, exclude)


print("Max profit from the projects:", budget_rec(costs, profit, budget, n))
```

```python
def budget_dp(costs, returns, budget, n, memo):
    if n == 0 or budget == 0:
        return 0
    
    if (n, budget) in memo:
        return memo[(n, budget)]
    
    if costs[n-1] > budget:
        memo[(n, budget)] = budget_dp(costs, returns, budget, n-1, memo)
    else:
        include = returns[n-1] + budget_dp(costs, returns, budget - costs[n-1], n-1, memo)
        exclude = budget_dp(costs, returns, budget, n-1, memo)
        memo[(n, budget)] = max(include, exclude)
    
    return memo[(n, budget)]

memo = {}
print("Max profit from the projects:", budget_dp(costs, returns, budget, n, memo))
```
